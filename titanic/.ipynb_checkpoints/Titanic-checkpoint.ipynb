```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn import preprocessing
import matplotlib.pyplot as plt
from sklearn.naive_bayes import MultinomialNB 
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv("train.csv")
```

```python
dataset
```

```python
def processFeatureSet(dataset):
    featureset = dataset.drop(['Name', 'Ticket', 'Cabin','Embarked'],axis = 1)
    encoder = preprocessing.LabelEncoder()
    encoder.fit(["female", "male"])
    featureset['Sex'] = encoder.transform(featureset['Sex'])
    featureset['Age'] = np.nan_to_num(featureset['Age'])
    return featureset
```

```python
dataset = shuffle(dataset)
```

```python
label = dataset['Survived']
```

```python
featureset = processFeatureSet(dataset)
featureset = featureset.drop('Survived', axis=1)
```

```python
featureset
```

```python
testSize = 891//10
```

```python
label_test,label_train = label[:testSize], label[testSize:]
```

```python
X_test,X_train = featureset[:testSize], featureset[testSize:]
```

```python
classifier = MultinomialNB()
classifier.fit(X_train, label_train)
y_predicted = classifier.predict(X_test)
print("Accuracy: %.2f" % accuracy_score(label_test, y_predicted))
```
